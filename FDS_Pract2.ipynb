# Line chart

A line chart is a graphical representation of an asset's historical price action that connects a series of data points with a continuous line. This is the most basic type of chart used in finance, and it typically only depicts a security's closing prices over time.

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12,7))
countries = ['USA','Brazil','Russia','Spain','UK','India']
totalDeaths = [112569,11247,587895,45786,158785,89745]
plt.bar(countries,totalDeaths, width=0.9,align='center',color='cyan',edgecolor='red')
i=1.0
j=2000

for i in range(len(countries)):
    plt.annotate(totalDeaths[i],(-0.1 + i, totalDeaths[i] + j))
plt.legend(labels = ['Total Deaths']) 
plt.title('Bar plot representing the total deaths by top 6 countries')
plt.xlabel('Countries')
plt.ylabel('Deaths')
plt.savefig('1BarPlot.png')
```

# stacking bar plots

```python
import pandas as pd
df = pd.DataFrame({
    'name':['john','mary','peter','ravina','jeff','bell','lisa','jose'],
    'age':[21,22,34,15,22,18,19,20],
    'gender':['M','M','F','M','F','M','F','F'],
    'state':['california','maharaashtra','shimla','dc','new york','dc','california','shimla'],
    'num_children':[2,0,1,0,3,1,2,1],
    'num_pets':[5,0,1,5,6,2,1,4]  
})

df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True)
```

```python


df.groupby(['gender','state']).size().unstack().plot(kind='bar',stacked=True)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[15,5])
countries= ['USA','Brazil','Russia','Spain','Uk','India']
totalCases = (20455,48789,48789,1565,5656,4879)
totalDeaths = (14521,7875,456,8954,5989,8995)
for i in range(len(countries)):
    plt.bar(countries[i],totalDeaths[i],bottom = totalCases[i]-totalDeaths[i],color="black")
    plt.bar(countries[i],totalCases[i] - totalDeaths[i], color='red')
    
    
plt.legend(labels = ['Total Deaths','Total Cases'])  
plt.title("Bar Plot representing the total deaths and total cases country wise")
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.savefig('3BarPlot.png')
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[15,10])
totalDeath=[11303,378894,48784,87545,475885]
totalRecovery=[7455,4845,5655,445,555]
activeCases=[1545,1545,1578,4989,48898]
country=['USA','India','Austria','Canada','Italy']
X = np.arange(len(totalDeath))
plt.bar(X, totalDeath, color='black',width=0.25)
plt.bar(X+0.25,totalRecovery,color='g',width=0.25)
plt.bar(X+0.5,activeCases,color='b',width=0.25)
plt.legend(['Total Deaths','Total Recovery','Active Cases'])
plt.xticks([i + 0.25 for i in range(5)],country)
plt.title("Bar plot representing the total Deaths , Total Recovery, Active Cases")
plt.xlabel('Countries')
plt.ylabel('Cases')
plt.savefig('4BarPlot.png')
plt.show()
```

# Pie chart

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
Data = {'Tasks':[300,400,200],
        'Task type':['Task pending','Task ongoing','Task Completed']
        }
df= DataFrame(Data)
df.set_index('Task type',inplace=True)
df.plot.pie(y='Tasks',figsize=(10,10),autopct='%1.1f%%', startangle=90)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,50,20,10]
fig,ax = plt.subplots()
ax.pie(sizes,labels=labels,autopct='%1.1f%%')
ax.axis('equal')
ax.set_title('Engineering Diciplines')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
labels = ['Civil','Electrical','Mechanical','Chemical']
sizes = [15,50,20,10]
explode = (0.1,0.1,0.1,0.4)
fig, ax = plt.subplots()
ax.pie(sizes,
      explode=explode,
      labels=labels,
      autopct='%1.1f%%',
      shadow=True,
      startangle=90)

ax.axis('equal')
ax.set_title('Engineering Diciplines')
plt.show()
```
